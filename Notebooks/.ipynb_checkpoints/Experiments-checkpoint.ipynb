```python
def plot_sims(dataset,x=2,y=2,tick=4,main_title="Moments for five ensemble runs",fig_size=[13,12]):
    sns.set_theme(style="darkgrid")
    var=["Lc","Nc","Lr","Nr"]
    
    fig= plt.figure()
    fig.set_size_inches(fig_size)
    c=["#26235b","#bc473a","#812878","#f69824","green"]
   
    for i in range(4):
        ax = fig.add_subplot(x,y, i + 1)
       
        for j in range (5):
            l=dataset[:,:,j]
            data=np.ma.compress_rows(l)
            data = np.delete(data, 3, 1)
            sns.lineplot(x=data[:,0],y=data[:,i+1],color=c[j],label="Sim No:"+str(j+1))
        plt.title(var[i])
      
        plt.xlabel("Time")
    
    fig.suptitle(f"Lo={dataset[1,-4,0]}; rm={dataset[1,-3,0]}; nu={dataset[1,-2,0]}", fontsize="x-large")
```

```python
import pytorch_lightning as pl
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from tqdm import tqdm
```

```python

class my_dataset(Dataset):
    def __init__(self,inputdata,updatedata,metadata):
        self.odddata=inputdata
        self.updatedata=updatedata
        self.metadata=metadata
               
    def __getitem__(self, index):
        return self.inputdata[index],self.updatedata[index],self.metadata[index]
    def __len__(self):
        return (self.odddata.shape)
```

```python
class DataModule(pl.LightningDataModule):

    def __init__(self, data_dir: str = None, 
                 batch_size: int = 256, num_workers: int = 0, transform=None):
        super().__init__()
        
        
    self.data_dir = data_dir
    self.batch_size = batch_size
    self.num_workers = num_workers
    self.transform = transform
    
    self.inputs=None
    self.updates=None
    self.outputs=None
    
    sel meta=None
    self.arr=None
    
    
    def setup:
        os.chdir(self.data_dir)
        with np.load('superdroplet_data_all.npz') as npz:
            self.arr = np.ma.MaskedArray(**npz)
        
        self.get_tendency()
        self.get_inputs()
        self.test_train()
        
        
        
        
        
    #Works well
    def get_tendency():
        arr_new=np.delete(self.arr, [3,6], 1)

        output_tend_all=[]
        for i in range (5):
            output_tend=[]
            for j in range(819):
                sim_data=arr_new[:,:,j,i]
                sim_data=np.ma.compress_rows(sim_data)
                output_sim=[]
                for t in range (1,len(sim_data)):
                    output_sim.append((sim_data[t,1:5]-sim_data[t-1,1:5])/20)

                output_tend.append(output_sim)

            output_tend_all.append(output_tend)


        self.updates,self.updates_mean,self.updates_std=self.norm(np.asarray(output_tend_all))


            
        
    def get_inputs():
        arr_new=np.delete(self.arr, [3,6], 1)

        input_all=[]
        meta_all=[]

        for i in range (5):
            inputs_sim=[]
            meta_sim=[]
            for j in range(819):
                sim_data=arr_new[:,:,j,i]
                sim_data=np.ma.compress_rows(sim_data)
                inputs=[]

                meta=sim_data[:-1,7:13]

                tau=sim_data[:-1,3]/(sim_data[:-1,3]+sim_data[:-1,1])
                xc=sim_data[:-1,1]/sim_data[:-1,2]

                inputs=np.concatenate((sim_data[:-1,1:5],tau.reshape(-1,1),xc.reshape(-1,1),sim_data[:-1,-4:-1]),axis=1)

                inputs_sim.append(inputs)
                meta_sim.append(meta)
            input_all.append(inputs_sim)
            meta_all.append(meta_sim)
        #call norm fucntion here and return to self values
        
        self.inputs,self.inputs_mean,self.inputs_std=self.norm(np.asarray(input_all))
        self.meta,self.meta_mean,self.meta_std=self.norm(np.asarray(meta_all))
        
            
    
    
    
    
    
    
    
    def norm(self,data):
        a=np.mean(input_all[0,:,:,:],axis=0)
        b_mean=np.mean(a,axis=0)

        a=np.std(input_all[0,:,:,:],axis=0)

        b_std=np.std(a,axis=0)

        new_data=(input_all-b_mean)/b_std

        return new_data,b_mean, b_std
    
    
    def test_train(self):
            test_len=219
            tot_len=self.inputs.shape[1]
            #These are exclusively for simulations
            
            self.sim_dataset_inputs=self.inputs[:,tot_len-test_len:,:,:]
            self.sim_dataset_meta=self.meta[:,tot_len-test_len:,:,:]
            self.sim_dataset_updates=self.updates[:,tot_len-test_len:,:,:]
            
            #reshaping needs to happen here
            
            self.reshape_everything(tot_len-test_len)
            
            dataset=my_dataset(self.inputs,self.meta,self.updates)
            
            validation_split = .2
      
            testing_split=.2
            shuffle_dataset = True


            # Creating data indices for training and validation splits:
            dataset_size = len(dataset)


            train_size = int(0.8 * dataset_size)
            test_size = dataset_size - train_size
            self.train_dataset, self.test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])
            train_size=int(0.9 * len(self.train_dataset))
            val_size=len(self.train_dataset)-train_size

            self.train_dataset, self.val_dataset = torch.utils.data.random_split(self.train_dataset, [train_size, val_size])
        
            
    
    
    def reshape_everything(self,train_len):
        input_reshaped=[]
        meta_reshaped=[]
        updates_reshaped=[]
        
        
        for i in range (5):
            for j in range (819):
                sim_data=self.inputs[i,j,:,:]
                sim_data=np.ma.compress_rows(sim_data)
                input_reshaped.append(sim_data)
                
                
                sim_data=self.meta[i,j,:,:]
                sim_data=np.ma.compress_rows(sim_data)
                meta_reshaped.append(sim_data)
                
                
                sim_data=self.updates[i,j,:,:]
                sim_data=np.ma.compress_rows(sim_data)
                updates_reshaped.append(sim_data)
                
                
                
        self.inputs=input_reshaped
        self.meta=meta_reshaped
        self.updates=updates_reshaped
        
        
        
    def train_dataloader(self):
        return DataLoader(self.train_dataset, batch_size=self.batch_size, num_workers=self.num_workers,
                          shuffle=True)

    def val_dataloader(self):
        return DataLoader(self.val_dataset, batch_size=self.batch_size, num_workers=self.num_workers,
                          shuffle=False)

    def test_dataloader(self):
        return DataLoader(self.test_dataset, batch_size=self.batch_size, num_workers=self.num_workers,
                          shuffle=False)

                
        
        
```
